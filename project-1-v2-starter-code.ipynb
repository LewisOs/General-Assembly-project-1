## 1. Defining a player

---

The player variables are:

    player_id : id code unique to each player (integer)
    player_name : entered name of the player (string)
    time_played : number of times played the game in minutes (float)
    player_pokemon: the player's captured pokemon (dictionary)
    gyms_visited: ids of the gyms that a player has visited (list)
    
Create the components for a player object by defining each of these variables. The dictionary and list variables should just be defined as empty; you can use any (correctly typed) values for the others.

````python
# creating Player class
class Player:
    """
    Represents a Pokemon Stay player
    
    ----------
    Attributes
    ----------
    player_id : int
        id code unique to each player
    player_name : str
        entered name of the player
    time_played : float
        amount of times played, in minutes
    player_pokemon: dict
        the player's captured pokemon
    gyms_visited : list
        names of the gyms the player has visited
    """
    def __init__(self, player_id, player_name, time_played = 0.0, player_pokemon = {}, gyms_visited = []):
        self.player_id = player_id
        self.player_name = player_name
        self.time_played = time_played
        self.player_pokemon = player_pokemon
        self.gyms_visited = gyms_visited
        
    def __str__(self):
        return self.player_name
    
    def get_name(self):
        """
        returns the player's name (player_name)
        """
        return self.player_name
````

````python
# creating player instance, ash
ash = Player(player_id = 1,
            player_name = 'Ash',
            time_played = 0.0,
            player_pokemon = {},
            gyms_visited = [],
            )
````

## 7. Calculate player "power".

---

Define a function that will calculate a player's "power". Player power is defined as the sum of the base statistics of all their pokemon.

Your function will:

1. Accept the `players` dictionary, `pokedex` dictionary, and a player_id as arguments.
2. For the specified player_id, look up that player's pokemon and their level(s).
3. Find and aggregate the attack and defense values for each of the player's pokemon from the `pokedex` dictionary.
4. Print "[player name]'s power is [player power].", where the player power is the sum of the base statistics for all of their pokemon.
5. Return the player's power value.

Print out the pokemon power for each of your players.

````python
def get_pokemon_strength(pokemon_id, pokedex = pokedex):
    """
    Calculates the strength of a pokemon:
    
    Takes a pokemon_id as input. Finds the pokemon associated with the pokemon_id in the pokedex dict.
    Sums the values for that pokemon's attack, defense, special_attack and special_defense attributes.
    Summed value is the pokemon's strength. Returns the summed value (int).
    
    ---------
    Arguments
    ---------
    pokemon_id : int
        id code unique to each pokemon
    pokedex : dict
        contains every instance of Pokemon. k = intance.pokemon_id, v = instance
    
    ---------
    Returns
    ---------
    pokemon_strength : int
        the sum of the pokemon's attack, defense, special_attack and special_defense attributes
    """
    pokemon_strength = sum([pokedex[pokemon_id].attack,
                           pokedex[pokemon_id].defense,
                           pokedex[pokemon_id].special_attack,
                           pokedex[pokemon_id].special_defense])
    
    return(pokemon_strength)


def get_player_power(player_id, players = players, pokedex = pokedex):
    """
    Calculates a player's power:
    
    Takes a player_id as input. Finds the player associated with that player_id in the players dict. Accesses
    that player's player_pokemon attribute to get a list of that player's pokemon. Calls the 
    get_pokemon_strength function on each pokemon in the list to create a list of each pokemon's strength.
    Sums the list of pokemon strengths. Summed value is the player's power. Returns the summed value (int).
    
    ---------
    Arguments
    ---------
    player_id : int
        id code unique to each player
    players : dict
        contains every instance of Player. k = instance.player_id, v = instance
    pokedex : dict
        contains every instance of Pokemon. k = intance.pokemon_id, v = instance
        
    --------
    Returns
    --------
    player_power_value : int
        the sum of the strength of every pokemon in the player's player_pokemon attribute
    """
    # get player's pokemon (list of pokemon_ids)
    players_pokemon_are = [pokedex[pokemon].pokemon_id for pokemon  in players[player_id].player_pokemon]
    
    # get pokemon strength (list of pokemon strengths)
    players_pokemons_strength = [get_pokemon_strength(pokemon) for pokemon in players_pokemon_are]
    
    # get player power
    player_power_value = sum(players_pokemons_strength)
    
    # printing
    print(f"{players[player_id].get_name()}'s power is {player_power_value}")
    
    return player_power_value


# printing players' power values
for player in players:
    get_player_power(player)
````

### 8.1

While you were putting together the prototype code, your colleagues were preparing a dataset of Pokemon and their attributes. (This was a rush job, so they may have picked some crazy values for some...)

The code below loads information from a comma separated value (csv) file. You need to parse this string into a more useable format. The format of the string is:

- Rows are separated by newline characters: \n
- Columns are separated by commas: ,
- All cells in the csv are double quoted. Ex: "PokedexNumber" is the first cell of the first row.


Using for-loops, create a list of lists where each list within the overall list is a row of the csv/matrix, and each element in that list is a cell in that row. Additional criteria:

1. Quotes are removed from each cell item.
2. Numeric column values are converted to floats.
3. There are some cells that are empty and have no information. For these cells put a -1 value in place.

Your end result is effectively a matrix. Each list in the outer list is a row, and the *j*th elements of the list together form the *j*th column, which represents a data attribute. The first three lists in your pokedex list should look like this:

    ['PokedexNumber', 'Name', 'Type', 'Total', 'HP', 'Attack', 'Defense', 'SpecialAttack', 'SpecialDefense', 'Speed']
    [1.0, 'Bulbasaur', 'GrassPoison', 318.0, 45.0, 49.0, 49.0, 65.0, 65.0, 45.0]
    [2.0, 'Ivysaur', 'GrassPoison', 405.0, 60.0, 62.0, 63.0, 80.0, 80.0, 60.0]

````python
# Code to read in pokedex info
raw_pd = ''
pokedex_file = 'pokedex_basic.csv'
with open(pokedex_file, 'r') as f:
    raw_pd = f.read()
    
# the pokedex string is assigned to the raw_pd variable
````

````python
# getting rows

# temp list to store rows
pd_rows = []

# looping through raw_pd's chars, appending chars between each '\n' to pd_rows
prev_row_end = 0
for i in range(len(raw_pd)):
    if raw_pd[i] == '\n':
        pd_rows.append(raw_pd[prev_row_end:i])
        prev_row_end = i+2
pd_rows.append(raw_pd[prev_row_end:])
````

````python
# getting cells


# list to store final cleaned version of pokedex
pd_final = []


# looping through pd_rows' elements
for row in range(len(pd_rows)):
    
    # add an empty row to pd_final
    pd_final.append([])
    # removing quotation marks from current row in pd_rows 
    pd_rows[row] = pd_rows[row].replace('"','')
    
    # splitting into cells
    prev_cell_end = 0
    for i in range(len(pd_rows[row])):
        if pd_rows[row][i] == ',':
            pd_final[row].append(pd_rows[row][prev_cell_end:i])
            prev_cell_end = i+1
    pd_final[row].append(pd_rows[row][prev_cell_end:])

 
                               
# changing empty cells to -1
for row in range(1,len(pd_final)):
    for i in pd_final[row]:
        if i == '':
            i = -1

# converting numeric columns to floats (skipping index, i.e. row 0)
for row in (pd_final[1:]):
    row[0] = float(row[0].lstrip('0'))
    for i in [3,4,5,6,7,8,9]:
        row[i] = float(row[i])
    

# Checking output is correct
example_above = [['PokedexNumber', 'Name', 'Type', 'Total', 'HP', 'Attack', 'Defense', 'SpecialAttack', 'SpecialDefense', 'Speed'],
                 [1.0, 'Bulbasaur', 'GrassPoison', 318.0, 45.0, 49.0, 49.0, 65.0, 65.0, 45.0],
                 [2.0, 'Ivysaur', 'GrassPoison', 405.0, 60.0, 62.0, 63.0, 80.0, 80.0, 60.0]]

for row in range(0,3):
    print(pd_final[row])

print(pd_final[0:3] == example_above)
````

## 12. Calibrate the frequency of Pokemon

The design team wants you to make the powerful Pokemon rare, and the weaklings more common. How would you set the probability $p_i$ of finding Pokemon *i* each time a player visits a gym?

Write a function that takes in a Pokedex number and returns a value $p_i$ for that character.

Hint: there are many ways you could do this. What do _you_ think makes sense? Start with simplifying assumptions: for example, you could assume that the probabilities of encountering any two Pokemon on one visit to a gym are independent of each other.

````python
"""
Assuming that the probabailities of encountering any two Pokemon on one visit from a gym are independent and that 
more powerful pokemon should be rarer, it is possible to set probabilities by:
    - taking each pokemon's total attribute to incidate their power & thus rareness
    - assigning an initial probability of 1 to all pokemon
    - dividing the probability of finding a pokemon by their power, such that the greater the power, the lower
    the probability.
"""

import pandas as pd
import matplotlib.pyplot as plt

def get_poke_prob(full_pokedex = full_pokedex):
    """
    Calculates the probability of finding all pokemon in the full_pokedex, given their power (i.e. total attribute)
    
    --------
    Arguments
    --------
    full_pokedex : dict
        contains every instance of Pokemon. k = intance.pokemon_id, v = instance
        
    --------
    Returns
    --------
    poke_probs : pandas.DataFrame
        rows = all pokemon in full_pokedex, cols = ['Total','Probability']
    """
    poke_probs_list = []   
    
    # getting name, total, probability
    for poke in full_pokedex:
        poke_list = []
        poke_list.append(full_pokedex[poke].name)
        poke_list.append(full_pokedex[poke].total)
        poke_list.append(10000 / full_pokedex[poke].total)
        poke_probs_list.append(poke_list)
    
    # creating and formatting Dataframe
    poke_probs = pd.DataFrame(poke_probs_list)
    poke_probs.columns = ['Name','Total','Probability']
    poke_probs.set_index('Name', inplace=True)
        
    return(poke_probs)

# getting pokemon probabilities 
probs = get_poke_prob()

# plotting pokemon probabilities to check output correct
plt.scatter(
    probs['Total'],
    probs['Probability'])
plt.title('Probability of finding a pokemon, as a function of pokemon total stat')
plt.xlabel('Pokemon total stat')
plt.ylabel('Probability of finding pokemon (%)')
````
