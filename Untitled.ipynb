```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime

sns.set(font_scale=1.5)
%config InlineBackend.figure_format = 'retina'
%matplotlib inline

from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression, Ridge, Lasso, ElasticNet, RidgeCV, LassoCV, ElasticNetCV
```

```python
police_2016 = pd.read_csv('/Users/lewis/Desktop/GA/DSI25-lessons/week04/day_1_/thecounted-data/the-counted-2016.csv')
police_2016.head()
```

```python
police = police_2015.append(police_2016)
```

police.month.unique()

```python
# dropping rows with missing values

orig_shape = police.shape
print(orig_shape)

# dropping unknown age
police = police[police.age != 'Unknown']
print(police.shape)

# dropping unknown raceethnicity
police = police[police.raceethnicity != 'Unknown']
print(police.shape)

# dropping unknown streetaddress
police = police[police.streetaddress != 'Unknown']
print(police.shape)

# dropping unknown lawenforcementagency
police = police[police.lawenforcementagency != 'Unknown']
print(police.shape)

# dropping unknown armed
police = police[police.armed != 'Unknown']
print(police.shape)

# dropping remaining NaNs
police.dropna(inplace = True)
print(police.shape)

print(f'{orig_shape[0]-police.shape[0]} rows of {orig_shape[0]} were dropped. {police.shape[0]} rows remaining.')
```

```python
(police.values == 'Unknown').sum()
```

```python
police.columns
```

```python
# updating & cleaning vars

# age
police.age[police.age == '40s'] = '45'
police.age = police.age.astype(float)



# agency
agency_col = []
for row in police.lawenforcementagency.values:
    if 'Sheriff' in row:
        agency_col.append('Sheriff')
    elif 'Police' in row:
        agency_col.append('Police')
    elif 'Veteran' in row:
        agency_col.append('Veteran Assoc')
    elif 'Corrections' in row:
        agency_col.append('Dept. Corrections')
    elif 'Highway' in row:
        agency_col.append('Highway Patrol')
    elif 'Public Safety' in row:
        agency_col.append('Dept. Public Safety')
    else: 
        agency_col.append('Other')
police.lawenforcementagency = pd.Series(agency_col)



# raceethnicity
police['is_POC'] = police['raceethnicity'] != 'White'



# armed
police['was_armed'] = police['armed'] == 'No'
police.rename(columns={'armed':'armed_with'})


# day month year to date time
days = [str(day) + ',' for day in police.day]
months = [str(month) + ',' for month in police.month]
years = [str(year) for year in police.year]

dates = []
for row in range(len(days)):
    date = days[row] + months[row] + years[row]
    dates.append(date)
    
police['datetime'] = [datetime.strptime(date,'%d,%B,%Y') for date in dates]
```

```python
police['was_armed'] = [not elem for elem in police.was_armed.values]
```

```python
dummies = pd.get_dummies(police[['gender', 'raceethnicity', 'month', 'state', 'classification','lawenforcementagency', 'armed']],drop_first=True)
```

```python
numerics = police[['age', 'is_POC', 'was_armed']]
```

```python
data = numerics.join(dummies)
data.columns
```

```python
y = data.age
X = data[data.columns[1:]]
```

```python
train_test_split(X,y)

ridge = RidgeCV(alphas=np.logspace(-5,5,100))
ridge.fit(X,y)
ridge.score(X,y)
```

```python
coefs = pd.Series(index = data.columns[1:],data = ridge.coef_).sort_values(ascending = False)
```

```python
fig, ax = plt.subplots(figsize=(30,10)) 
coefs.plot.bar()
plt.ylabel('Coefficient',fontsize = 30)
plt.xlabel('Variable',fontsize = 30)
plt.title('Variables by coefficient size',fontsize = 50)
```
